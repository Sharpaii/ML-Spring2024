#### Part 1 data preparation

In the cell below, generate training data for your model.
<br> Generate a decent amount of training data in the interval of x.
<br>
<br>All data need to be separated by the same distance.


```python
import matplotlib.pyplot as plt
import numpy as np

xc = np.linspace(-100, 100, 100000)
yc = xc * np.sin(xc ** 2 / 300)

plt.plot(xc, yc)
plt.show()
```

#### Part 2 build models

In the cells below, build your models and train it with data from part1.
<br>You need to split your training data to two parts. With 40% used in training, and 60% used in test.

```python
# model 3
from tensorflow import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout, Activation, Flatten
from sklearn.model_selection import train_test_split

# Split the data
x, x_test, y, y_test = train_test_split(xc, yc, test_size=0.15, shuffle=True)

# split remaining datato train & valid
X_train, x_val, Y_train, y_val = train_test_split(x, y, test_size=0.40, shuffle=True)

model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(150,activation = "relu"))
model.add(Dense(150,activation = "LeakyReLU"))
model.add(Dense(150,activation = "ThresholdedReLU"))
model.add(Dense(150,activation = "PReLU"))
model.add(Dense(150,activation = "ELU"))
model.add(Dense(150,activation = "relu"))
model.add(Dense(1))

model.compile(
   loss = 'MAE', optimizer = 'Adam', metrics=['mae']
)

model.fit(x=X_train, y=Y_train, batch_size=64, epochs=100, validation_data=(x_val, y_val))

```

#### Part 3 model evaluation

In the cells below, eval your model with training data, test data ( with model.evaluate( ) ), and plot your predict result on the same plot with the plot of goal function.

```python
py = model.predict(xc)

plt.plot(xc, py)
plt.plot(xc, yc)
plt.title("Model Fit")
plt.show()
es = []
for i in range(len(py)):
  es.append(yc[i] - py[i])
plt.plot(xc, es)
plt.title("Error")
plt.show()
```
