```python
weo_raw = pd.read_csv("C:/Users/madel/Downloads/WEOOct2023all.xls", sep='\t', encoding='ANSI')
print(weo_raw)

print(weo_raw.columns.tolist())
```

```python
weo_selected_measurement = weo_raw.loc[weo_raw['WEO Subject Code'].str.contains("NGDP_RPCH", na=False)]
weo_selected_measurement_2018 = pd.DataFrame(weo_selected_measurement, columns=['Country', '2018'])

print(weo_selected_measurement_2018)
```

#### Clean Data for Training

```python
merged_train_data_1 = pd.merge(hls_train_1, weo_selected_measurement_2018, on="Country")
merged_train_data_1 = merged_train_data_1.rename(columns={"Value": "Difficulty Making Ends Meet Measurement", "2018": "Income Measurement"})
merged_train_data_1 = pd.DataFrame(merged_train_data_1, columns=['Country','Difficulty Making Ends Meet Measurement', 'Income Measurement'])

print(merged_train_data_1)
```

```python
merged_train_data_2 = pd.merge(hls_train_2, weo_selected_measurement_2018, on="Country")
merged_train_data_2 = merged_train_data_2.rename(columns={"Value": "Access to Green Space Measurement", "2018": "Income Measurement"})
merged_train_data_2 = pd.DataFrame(merged_train_data_2, columns=['Country','Access to Green Space Measurement', 'Income Measurement'])

print(merged_train_data_2)
```

```python
merged_train_data_3 = pd.merge(hls_train_3, weo_selected_measurement_2018, on="Country")
merged_train_data_3 = merged_train_data_3.rename(columns={"Value": "Housing Cost Overburden Measurement", "2018": "Income Measurement"})
merged_train_data_3 = pd.DataFrame(merged_train_data_3, columns=['Country','Housing Cost Overburden Measurement', 'Income Measurement'])

print(merged_train_data_3)
```

### Fit Model with Data

```python
import matplotlib.pyplot as plt
import sklearn.linear_model
from sklearn.metrics import mean_absolute_error

X = np.c_[merged_train_data_1["Income Measurement"]]
Y = np.c_[merged_train_data_1["Difficulty Making Ends Meet Measurement"]]
x = X.tolist()
y = Y.tolist()

# plot data
out1 = widgets.Output()
with out1:
  plt.scatter(x, y)
  plt.xlabel('Income')
  plt.ylabel('Difficulty Making Ends Meet')
  plt.title("Data Plot")
  plt.show()

# fit linear model
model = sklearn.linear_model.LinearRegression()
model.fit(X, Y)

# plot predictions
predict_x = [x for x in range(901)]
predict_x = [[x/100] for x in predict_x]
predict_y = model.predict(predict_x)

out2 = widgets.Output()
with out2:
  plt.scatter(predict_x, predict_y)
  plt.scatter(x, y)
  plt.xlabel('Income')
  plt.ylabel('Difficulty Making Ends Meet')
  plt.title("Prediction Line")
  plt.show()

display(widgets.HBox([out1,out2]))


# Calculate MAE
predictions = model.predict(X)
mae = mean_absolute_error(Y, predictions)
print("Mean Absolute Error: ", mae)
```
